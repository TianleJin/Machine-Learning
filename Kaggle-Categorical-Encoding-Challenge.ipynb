```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('./train.csv')
test_df = pd.read_csv('./test.csv')
sample_sub_df = pd.read_csv('./sample_submission.csv')
```

```
# df labels
labels = [col for col in train_df.columns if col != 'target']
print(f'df labels: {labels}')
```

```
# drop unwanted columns
combined = [train_df, test_df]
unwanted = ['nom_5', 'nom_6', 'nom_7', 'nom_8', 'nom_9']
for df in combined:
  for title in unwanted:
    df.drop(title, axis=1, inplace=True)
    
train_df.head()
```

```
# fill NaN with 0s in bin_0 and bin_1 due to high percentage of 0s
train_df[['bin_0', 'bin_1']].describe(percentiles=[.775, .8, .85, .9], include=np.number)
```

```
test_df[['bin_0', 'bin_1']].describe(percentiles=[.775, .8, .85, .9], include=np.number)
```

```
# forward fill null entries
for df in combined:
  df.fillna(method='bfill', inplace=True)

train_df.info()
```

```
# nom_0 - nom_4 categorical encoding
for c in ['nom_0', 'nom_1', 'nom_2', 'nom_3', 'nom_4']:
  print(f'train_df[{c}]: {pd.unique(train_df[c])}')
```

```
# ord_1 - ord_5 normal encoding
ords = ['ord_0', 'ord_1', 'ord_2', 'ord_3', 'ord_4']
for category in ords:
  print(train_df[category].value_counts())
```

```
encode_labels = ['bin_3', 'bin_4', 'nom_0', 'nom_1', 'nom_2', 'nom_3', 'nom_4', 'ord_1', 'ord_2', 
                 'ord_3', 'ord_4', 'ord_5']
encode = []
for c in encode_labels:
  encode.append(dict((ty, num) for num, ty in enumerate(pd.unique(train_df[c]))))
```

```
for c, en in zip(encode_labels, encode):
  train_df[c] = train_df[c].apply(lambda x:en[x])
  test_df[c] = test_df[c].apply(lambda x:en[x])

train_df.info()
```

```
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
```

<h2>Logistic Regression Model</h2>

```
x_train = train_df.drop('target', 1).drop('id', 1).values
y_train = train_df['target'].values 
```

```
clf = LogisticRegression(solver='liblinear')
cv_results = cross_validate(clf, x_train, y_train)
```

```
cv_results['test_score']
```

```
clf.fit(x_train, y_train)
```

<h2>Submission</h2>

```
x_test = test_df.drop('id', 1).values
```

```
y_hat = clf.predict_proba(x_test)
print(y_hat[0:5])
```

```
submission_lgr = pd.DataFrame()
submission_lgr['id'] = test_df['id']
submission_lgr['target'] = y_hat[:, 1]
submission_lgr.head(10)
```

```
submission_lgr.to_csv('./lgr_predictions.csv') 
```
